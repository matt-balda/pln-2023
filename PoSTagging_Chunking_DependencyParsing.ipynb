# Identificação de Part-of-Speech Tagging (POS)

```python
doc = nlp('A rápida raposa marrom saltou sobre o preguiçoso cachorro.')

for token in doc:
  print(token.text, "|", token.tag_, "|", spacy.explain(token.tag_))
```

```python
displacy.render(doc, style="dep", jupyter=True)
```

# Chucking e grupos nominais, verbais e etc...

```python
def print_chunks(doc):
    for chunk in doc.noun_chunks:
        print(f'Grupo Nominal: {chunk.text}')

    for chunk in doc:
        if 'VERB' in chunk.pos_:
            print(f'Grupo Verbal: {chunk.text}')

    for chunk in doc:
        if 'ADJ' in chunk.pos_:
            print(f'Grupo Adjetival: {chunk.text}')

    for chunk in doc:
        if 'ADV' in chunk.pos_:
            print(f'Grupo Adverbial: {chunk.text}')

    for chunk in doc:
        if 'PRON' in chunk.pos_:
            print(f'Grupo Pronominal: {chunk.text}')

    for chunk in doc:
        if 'ADP' in chunk.pos_:
            print(f'Grupo Prepositivo: {chunk.text}')

    for chunk in doc:
        if 'CONJ' in chunk.pos_:
            print(f'Grupo Conjunção: {chunk.text}')

    for chunk in doc:
        if 'DET' in chunk.pos_:
            print(f'Grupo Determinante: {chunk.text}')
```

```python
print_chunks(doc)
```
